```python
# 1.The Receipt Generator
# Create variables for three items: item1_price, item2_price, and item3_price.
# Ask the user to input the price of each item
# Calculate the subtotal.
# Calculate the tax (assume 8%).
# Calculate the total.
# Print a formatted receipt using F-strings
item1_price = float(input("Enter item 1 price: "))
item2_price = float(input("Enter item 2 price: "))
item3_price = float(input("Enter item 3 price: "))

subtotal = item1_price + item2_price + item3_price
tax = subtotal * 0.08
total = subtotal + tax

print(f"Subtotal: {subtotal}, Tax: {tax}, Total: {total}")
```

```python
# 3. String Slicer
# Given the string: s = "PythonProgramming".
# Write code to print:
# "Python"
# "gram"
# "gnimmargorPnohtyP" (The whole string reversed).
s = "PythonProgramming"

# 1. "Python" → first 6 characters
print(s[:6])

# 2. "gram" → characters from index 10 to 13
print(s[9:13])

# 3. Reverse the string → step -1
print(s[::-1])
```

```python
# 4. The modulus Check Write a script that defines a variable number. 
# Check if the number is even or odd using the % operator and print a
# boolean result (True if even, False if odd).
number = int(input("Enter a number"))
is_odd = (number %2!=0)
print(is_odd)
```

```python
# Start with this list: guests = ["Alice", "Bob", "Charlie", "David"].
# Perform the following steps in order:
# "Bob" cancels. Remove him.
# "Eve" is a VIP. Add her to the start of the list.
# "Frank" joins. Add him to the end of the list.
# Sort the list alphabetically.

# Print the final list and the total number of guests.
guests = ["Alice", "Bob", "Charlie", "David"]

guests.remove("Bob")

guests.insert(0, "Eve")

guests.append("Frank")

print("Final guest list:", guests)
print("Total number of guests:", len(guests))
```
