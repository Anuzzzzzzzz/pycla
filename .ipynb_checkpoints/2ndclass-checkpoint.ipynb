```python
# 3. String Slicer
# Given the string: s = "PythonProgramming".
# Write code to print:
# "Python"
# "gram"
# "gnimmargorPnohtyP" (The whole string reversed).
s = "PythonProgramming"

# 1. "Python" → first 6 characters
print(s[:6])

# 2. "gram" → characters from index 10 to 13
print(s[9:13])

# 3. Reverse the string → step -1
print(s[::-1])
```

```python
# Start with this list: guests = ["Alice", "Bob", "Charlie", "David"].
# Perform the following steps in order:
# "Bob" cancels. Remove him.
# "Eve" is a VIP. Add her to the start of the list.
# "Frank" joins. Add him to the end of the list.
# Sort the list alphabetically.

# Print the final list and the total number of guests.
guests = ["Alice", "Bob", "Charlie", "David"]

guests.remove("Bob")

guests.insert(0, "Eve")

guests.append("Frank")

print("Final guest list:", guests)
print("Total number of guests:", len(guests))
```

```python
# 8. The Duplicate Destroyer
# You have a list of ID numbers with duplicates: ids = [101, 102, 101, 103, 104, 102].

# Using one specific data structure conversion, create a new variable unique_ids that contains only unique numbers. 
# Print the count of unique IDs.
ids = [101, 102, 101, 103, 104, 102]
unique_elements=set(ids)
print(unique_elements)
count=len(unique_elements)
print(f"Count of Unique IDs: {count}")
```

```python
# Given the following complex dictionary representing a student database:

# school_data = {
#     "class_10A": {
#         "students": [
#             {"name": "John", "grades": {"math": 80, "science": 90}},
#             {"name": "Sarah", "grades": {"math": 95, "science": 88}}
#         ]
#     }
# }



# output:
# John: maths:80, science:90
# Sarah: maths:95, science:88


school_data = {
    "class_10A": {
        "students": [
            {"name": "John", "grades": {"math": 80, "science": 90}},
            {"name": "Sarah", "grades": {"math": 95, "science": 88}}
        ]
    }
}

students = school_data["class_10A"]["students"]

for student in students:
    name = student["name"]
    math = student["grades"]["math"]
    science = student["grades"]["science"]
    print(f"{name}: maths:{math}, science:{science}")
```

```python
# 10.The Data Parser(String Slicing)
# You have a date in a specific format: date_string = "2023-12-25".

# Use  (using square brackets []) to extract the year, month, and day into separate variables.
# year should be "2023"
# month should be "12"
# day should be "25"
#  Print them in US format: 12/25/2023 using an f-string.

date_string = "2023-12-25"
year= date_string[:4]
month= date_string[5:7]
day= date_string[8:10]
print(f"{month}/{day}/{year}")
```

```python
# The File Name Standardizer(Method Chaining)
# Users often upload files with messy names. You need to standardize them for a database.
#  file_name = " My Summer Vacation PHOTO.JPG "
#  Write a chain of methods to:
#  "my_summer_vacation_photo.jpg"

file_name = "My Summer Vacation PHOTO.jpg"
print(file_name.strip().lower().replace(" ","_"))
```
